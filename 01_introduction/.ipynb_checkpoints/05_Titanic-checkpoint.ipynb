# 타이타닉 생존자 예측

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../00_data/titanic/train.csv')
df
```

```python
df.info()
```

### 의미 없는 컬럼 제거 및 결측치 처리

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna('N',inplace=True)
```

```python
df.info()
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

for feature in ['Sex', 'Embarked']:
    encoder = LabelEncoder()
    df[feature] = encoder.fit_transform(df[feature])
```

```python
df
```

```python
df.info()
```

### 'Survived' 속성을 y로 나머지 feature를 X로 만듦

```python
y_df = df['Survived']
X_df = df.drop('Survived', axis=1)
```

### train/test 데이터 세트 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(
    X_df, y_df, test_size=0.2, random_state=11
)
```

### 학습/예측/평가

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# DecisionTreeClassifier 학습/예측/평가
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)
print(f'DecisionTreeClassifier 정확도: {accuracy_score(y_test, pred):.4f}')
```

### 교차 검증 (Stratified K-Fold)

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5)
scores = []
    
# KFold 교차 검증 수행. 
for iter_count, (train_index, test_index) in enumerate(skf.split(X_df, y_df)):
    # X_titanic_df 데이터에서 교차 검증별로 학습과 검증 데이터를 가리키는 index 생성
    X_train, X_test = X_df.values[train_index], X_df.values[test_index]
    y_train, y_test = y_df.values[train_index], y_df.values[test_index]
    
    # Classifier 학습, 예측, 정확도 계산 
    dt_clf.fit(X_train, y_train) 
    pred = dt_clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    scores.append(accuracy)
    print(f"교차 검증 {iter_count}  정확도: {accuracy:.4f}")     

# 5개 fold에서의 평균 정확도 계산. 
mean_score = np.mean(scores)
print(f"평균 정확도: {mean_score:.4f}")
```

### GridSearchCV를 통해 최적 파라미터 도출

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth':[2, 3, 5, 10],
    'min_samples_leaf':[1, 5, 8],
    'min_samples_split':[2, 3, 5]
}
grid_dtree = GridSearchCV(dt_clf, param_grid=parameters, scoring='accuracy', cv=5)
grid_dtree.fit(X_train, y_train)

print('GridSearchCV 최적 하이퍼 파라미터:', grid_dtree.best_params_)
print(f'GridSearchCV 최고 정확도: {grid_dtree.best_score_:.4f}')
best_dtree = grid_dtree.best_estimator_

# GridSearchCV의 최적 하이퍼 파라미터로 학습된 Estimator로 예측 및 평가 수행. 
pred = best_dtree.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f'테스트 세트에서의 DecisionTreeClassifier 정확도: {accuracy:.4f}')
```
